```python
f,ax = plt.subplots(1,2,figsize=(18,8))

data['Fatalities'].value_counts().plot.pie(autopct='%1.1f%%',ax=ax[0],shadow=True)

ax[0].set_title('Fatalities')
ax[0].set_ylabel('')


sns.countplot(x='Fatalities',data=data,ax=ax[1])
ax[1].set_title('Fatalities')
plt.show()
```

```python
data.groupby(['Aircraft_Type','Is_Fatal'])['Is_Fatal'].count()
```

```python
f,ax = plt.subplots(1,2,figsize=(18,8))
data[['Aircraft_Type','Is_Fatal']].groupby(['Aircraft_Type']).mean().plot.bar(ax=ax[0])
ax[0].set_title('is_fatal vs aircraft_type')

sns.countplot(x='Aircraft_Type',hue='Is_Fatal',data=data,ax=ax[1])
ax[1].set_title('Aircraft_Type:is_fatal va dead')
plt.show()
```

```python
pd.crosstab(data.,data.Injuries,margins=True).style.background_gradient(cmap='winter_r')
```

```python
f,ax=plt.subplots(1,2,figsize=(18,8))
data['Is_Fatal'].value_counts().plot.bar(color=['#CD7F32','#FFDF00'],ax=ax[0])
ax[0].set_title('number of injuris')
ax[0].set_ylabel('count')

sns.countplot(x='Injuries',hue='Is_Fatal',data=data,ax=ax[1])
ax[1].set_title('injuries:')

plt.show()
```

```python
pd.crosstab([data.Aircraft_Type,data.Is_Fatal],data.Injuries,margins=True).style.background_gradient(cmap='summer_r')
```

```python
sns.catplot(x='Injuries',y='Is_Fatal',hue='Aircraft_Type',data=data)
plt.show()
```
